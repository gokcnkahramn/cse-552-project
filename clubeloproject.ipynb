```python
# URL
url = "http://api.clubelo.com/Huddersfield"

# CSV çek
df = pd.read_csv(url)

# Script'in çalıştığı klasörü al
base_path = os.getcwd()
output_folder = os.path.join(base_path, "data", "clubelo")

# Klasör yoksa oluştur
os.makedirs(output_folder, exist_ok=True)

# Dosya yolu
file_path = os.path.join(output_folder, "clubelo_huddersfield.csv")

# CSV olarak kaydet
df.to_csv(file_path, index=False)

print(f"Saved")
```

```python
input_folder = os.path.join(base_path, "data", "clubelo")
output_file = os.path.join(input_folder, "clubelo.csv")

# Klasördeki tüm csv dosyalarını bul
csv_files = [f for f in os.listdir(input_folder) if f.endswith(".csv")]

dfs = []

for file in csv_files:
    file_path = os.path.join(input_folder, file)
    df = pd.read_csv(file_path)
    dfs.append(df)

# Hepsini birleştir
full_df = pd.concat(dfs, ignore_index=True)

# From kolonunu datetime yap
full_df["From"] = pd.to_datetime(full_df["From"], errors="coerce")

# 2017 ve 2018 için filtre
filtered = full_df[full_df["From"].dt.year.isin([2017, 2018])]

# Kaydet
filtered.to_csv(output_file, index=False)

print("clubelo.csv oluşturuldu.")
```
